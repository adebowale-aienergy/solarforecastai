```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming train_lstm and test_lstm are already defined and scaled

# Define a function to create sequences for LSTM
def create_sequences(data, sequence_length):
    """Create sequences from a time series data."""
    sequences = []
    targets = []
    for i in range(len(data) - sequence_length):
        sequences.append(data[i : i + sequence_length])
        targets.append(data[i + sequence_length])
    return np.array(sequences), np.array(targets)

sequence_length = 30 # Use the same sequence length as in the data preparation step

X_train_lstm = []
y_train_lstm = []
X_test_lstm = []
y_test_lstm = []

# Group data by location and create sequences for training and testing
grouped_train = train_lstm.groupby(['country', 'lat', 'lon'])
grouped_test = test_lstm.groupby(['country', 'lat', 'lon'])

for name, group in grouped_train:
    scaled_values = group['value_scaled'].values
    sequences, targets = create_sequences(scaled_values, sequence_length)
    if sequences.shape[0] > 0:
        X_train_lstm.append(sequences)
        y_train_lstm.append(targets)

for name, group in grouped_test:
    scaled_values = group['value_scaled'].values
    sequences, targets = create_sequences(scaled_values, sequence_length)
    if sequences.shape[0] > 0:
        X_test_lstm.append(sequences)
        y_test_lstm.append(targets)

# Stack sequences and targets
X_train_lstm = np.concatenate(X_train_lstm) if X_train_lstm else np.array([])
y_train_lstm = np.concatenate(y_train_lstm) if y_train_lstm else np.array([])
X_test_lstm = np.concatenate(X_test_lstm) if X_test_lstm else np.array([])
y_test_lstm = np.concatenate(y_test_lstm) if y_test_lstm else np.array([])

# Reshape for LSTM input [samples, time steps, features]
X_train_lstm = X_train_lstm.reshape((X_train_lstm.shape[0], X_train_lstm.shape[1], 1))
X_test_lstm = X_test_lstm.reshape((X_test_lstm.shape[0], X_test_lstm.shape[1], 1))

# Display the shapes of the prepared LSTM data
print("LSTM Training Data Shape:", X_train_lstm.shape)
print("LSTM Training Target Shape:", y_train_lstm.shape)
print("LSTM Testing Data Shape:", X_test_lstm.shape)
print("LSTM Testing Target Shape:", y_test_lstm.shape)

# Define the LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(units=50, return_sequences=False, input_shape=(sequence_length, 1)))
lstm_model.add(Dense(units=1))

# Compile the model
lstm_model.compile(optimizer='adam', loss='mse')

# Train the LSTM model
history = lstm_model.fit(X_train_lstm, y_train_lstm, epochs=10, batch_size=32, validation_split=0.2)

print("LSTM model defined, compiled, and trained.")

# Save the trained LSTM model
lstm_model_filename = 'lstm_model.h5'
lstm_model.save(lstm_model_filename)
print(f"LSTM model saved successfully as {lstm_model_filename}")
```
